```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import  seaborn as sns
import plotly.express as px
from sklearn.datasets import make_blobs

%matplotlib inline
```

```python
X,y= make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
## Standardization ----- Feature scaling techique
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
## train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow method to select kvalue 
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## plot elbow curve 
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of clusters ")
plt.ylabel("wcss")
```

```python
kmeans=KMeans(init="k-means++",n_clusters=3)
```

```python
kmeans.fit_predict(X_train_scaled)
```

```python
y_pred= kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

```python
## Validating the k value
## 1 kneelocator
## 2 silhoutee scoring
```

```python
from kneed import KneeLocator

kl = KneeLocator(range(1, 11), wcss, direction="decreasing", curve="convex")
print("Best number of clusters:", kl.knee)

```

```python
## silhoutte scoring 
from sklearn.metrics import silhouette_score
silhouette_coefficient=[]
for k in range (2,11):
    kmeans= KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    score=silhouette_score(X_train_scaled,kmeans.labels_)
    silhouette_coefficient.append(score)
```

```python
silhouette_coefficient
```

```python
plt.plot(range(2,11),silhouette_coefficient)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters ")
plt.ylabel("silhoutte coefficients")
```
